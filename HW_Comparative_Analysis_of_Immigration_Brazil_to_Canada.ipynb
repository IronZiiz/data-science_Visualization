```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/drive/MyDrive/Alura/Data_Visualization:criando-gráficos-com-bibliotecas-Python/DV-Immigrants-project/imigrantes_canada.csv')
```

```python
df.head()
```

```python
# Set the 'País' (Country) column as the index of the DataFrame.
# This makes it easier to access data based on the country.
df.set_index('País', inplace = True)
```

```python
# Create a list of years from 1980 to 2013 as strings.
# These years will be used to select specific columns from the DataFrame.
anos = list(map(str, range(1980, 2014)))
```

```python

# Extract the immigration data for Brazil for the years specified in the 'anos' list.
brasil = df.loc['Brasil', anos]

# Extract the immigration data for Argentina for the years specified in the 'anos' list.
argentina = df.loc['Argentina',anos]
```

```python
brasil
```

```python
argentina
```

```python
# Create a dictionary to store the year and the number of immigrants for Brazil.
brasil_dict = {'ano': brasil.index.tolist(), 'imigrantes': brasil.values.tolist()}

# Convert the 'brasil_dict' dictionary into a pandas DataFrame.
dados_brasil = pd.DataFrame(brasil_dict)
```

```python
dados_brasil
```

```python
# Create a dictionary to store the year and the number of immigrants for Argentina.
# .values extracts the numerical data from the Series and .tolist() converts it into a list.
argentina_dict = {'ano': argentina.index.tolist(), 'imigrantes': argentina.values.tolist()}

# Convert the 'argentina_dict' dictionary into a pandas DataFrame.
dados_argentina = pd.DataFrame(argentina_dict)
```

```python
dados_argentina
```

```python
plt.figure(figsize=(12, 6))  # Increased figure size for better readability

plt.plot(anos, brasil, label='Brasil')
plt.plot(anos, argentina, label='Argentina')

plt.legend(title='País de Origem', fontsize='large') #Added a title for the legend
plt.title('Imigração para o Canadá: Brasil X Argentina (1980-2013)', fontsize=14) #Improved the title
plt.xlabel('Ano', fontsize=12)
plt.ylabel('Número de Imigrantes', fontsize=12)
plt.xticks(['1980','1985','1990','1995','2000','2005', '2010'])
plt.grid(True, linestyle='--', alpha=0.7) #Added a grid for better visualization

plt.tight_layout() # Adjust layout to prevent labels from overlapping
plt.show()
```
