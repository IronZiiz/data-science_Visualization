```python
import pandas as pd
```

```python
df = pd.read_csv('/content/drive/MyDrive/Alura/Data_Visualization:criando-gráficos-com-bibliotecas-Python/DV-Immigrants-project/imigrantes_canada.csv')
```

```python
df.head()
```

```python
df.set_index('País', inplace = True)
```

```python
anos = list(map(str, range(1980, 2014)))
```

```python
anos
```

```python
 brasil = df.loc['Brasil', anos]
```

```python
brasil
```

```python
brasil_dict = {'ano': brasil.index.tolist(), 'imigrantes': brasil.values.tolist()}
dados_brasil = pd.DataFrame(brasil_dict)
```

```python
dados_brasil
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(dados_brasil['ano'], dados_brasil['imigrantes'])
plt.show()
```

```python
# improving visualization
plt.figure(figsize = (8, 4)) # inches
plt.plot(dados_brasil['ano'], dados_brasil['imigrantes'])
plt.title('Imigração do Brasil para o Canadá')
plt.xticks(['1980', '1990', '2000', '2010'])
plt.xlabel('Ano')
plt.ylabel('Número de Imigrantes')
plt.show()
```
