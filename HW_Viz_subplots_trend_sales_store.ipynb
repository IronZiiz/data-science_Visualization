```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# make a Df

lojas = ['A', 'B', 'C', 'D']

vendas_2022 = {'Jan': [100, 80, 150, 50],
    'Fev': [120, 90, 170, 60],
    'Mar': [150, 100, 200, 80],
    'Abr': [180, 110, 230, 90],
    'Mai': [220, 190, 350, 200],
    'Jun': [230, 150, 280, 120],
    'Jul': [250, 170, 300, 140],
    'Ago': [260, 180, 310, 150],
    'Set': [240, 160, 290, 130],
    'Out': [220, 140, 270, 110],
    'Nov': [400, 220, 350, 190],
    'Dez': [300, 350, 400, 250]
}
```

```python
dados_dict = {'Lojas': lojas}
dados_dict.update(vendas_2022)
```

```python
df = pd.DataFrame(dados_dict)
```

```python
df.head()
```

```python
df = df.set_index('Lojas')
df.head()
```

```python
fig, axs = plt.subplots(2,2, figsize = (12, 6))
fig.subplots_adjust(hspace=0.5 , wspace=0.3)
fig.suptitle('Vendas de cada loja por mês (2022)', fontsize = 16)

axs[0, 0].plot(df.loc['A', 'Jan':'Dez'])
axs[0, 0].set_title( 'Loja A', fontdict={'fontweight' : 'bold'})


axs[0, 1].plot(df.loc['B', 'Jan':'Dez'])
axs[0, 1].set_title('Loja B', fontdict={'fontweight' : 'bold'})


axs[1, 0].plot(df.loc['C', 'Jan':'Dez'])
axs[1, 0].set_title('Loja C', fontdict={'fontweight' : 'bold'})


axs[1, 1].plot(df.loc['D', 'Jan':'Dez'])
axs[1, 1].set_title('Loja D', fontdict={'fontweight' : 'bold'})

# adjust scales
ymin = 0
ymax = 500

for ax in axs.flat:
    ax.set(xlabel='Mês', ylabel='Vendas')
    ax.set(ylim=(ymin, ymax))
    ax.grid()

plt.show()
```

```python
# Comparative trend sales on the same grapg
plt.figure(figsize=(10, 6))
for loja in lojas:
    plt.plot(df.loc[loja, 'Jan':'Dez'], label=loja)

plt.title('Vendas de cada loja por mês (2022)', fontsize=16)
plt.xlabel('Mês')
plt.ylabel('Vendas')
plt.ylim(0, 500)  # Adjust y-axis limits
plt.grid(True)
plt.legend()  # Add a legend to differentiate the stores
plt.show()
```
