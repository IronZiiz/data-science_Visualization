```python
import pandas as pd
import plotly.express as px
import plotly.graph_objs as go
```

```python
df = pd.read_csv('/content/drive/MyDrive/Alura/Data_Visualization:criando-gráficos-com-bibliotecas-Python/DV-Immigrants-project/imigrantes_canada.csv')
```

```python
df.set_index('País', inplace = True)
```

```python
anos = list(map(str, range(1980, 2014)))
```

```python
america_sul = df.query('Região == "América do Sul"')
america_sul.head()
```

```python
df_america_sul_clean = america_sul.drop(['Continente', 'Região', 'Total'], axis = 1)
america_sul_final = df_america_sul_clean.T
```

```python
america_sul_final
```

```python
brasil_argentina = america_sul_final[['Brasil', 'Argentina']]
brasil_argentina
```

```python
fig = px.line(brasil_argentina, x = brasil_argentina.index, y = brasil_argentina.columns, color = 'País',
              title = 'Imigração dos países da América do Sul para o Canadá 1980\n 2013', markers = False)

fig.update_layout(
    xaxis_title = 'Ano',
    yaxis_title = 'Número de Imigrantes',
    font_family = 'Arial',
    font_color = 'grey',
    font_size = 14,
    title_font_color = 'black',
    title_font_size = 22,
)
```

```python
# Convertendo o índice para inteiro
brasil_argentina.index = brasil_argentina.index.astype(int)

fig = go.Figure()

# Adicionando um trace para cada país (iniciando com o primeiro ponto)
for col in brasil_argentina.columns:
    fig.add_trace(
        go.Scatter(
            x=[brasil_argentina.index[0]],
            y=[brasil_argentina[col].iloc[0]],
            mode='lines',
            name=col,
            line=dict(width=4)
        )
    )

# Configurando o layout do gráfico
fig.update_layout(
    title=dict(
        text='<b>Imigração do Brasil e Argentina para o Canadá (1980 a 2013)</b>',
        x=0.12,
        xanchor='left',
        font=dict(size=20)
    ),
    xaxis=dict(range=[1980, 2013], autorange=False, title='<b>Ano</b>'),
    yaxis=dict(range=[0, brasil_argentina.max().max() + 100], autorange=False, title='<b>Número de Imigrantes</b>'),
    updatemenus=[dict(
        type='buttons',
        showactive=False,
        buttons=[dict(
            label='Play',
            method='animate',
            args=[None, {'frame': {'duration': 100, 'redraw': True}, 'fromcurrent': True}]
        )]
    )],
    width=1000,
    height=500
)

# Criando os frames para a animação
frames = []
for i in range(1, len(brasil_argentina)):
    # Para cada frame, cada país terá os dados até o ano correspondente (índice i)
    frame_data = []
    for col in brasil_argentina.columns:
        frame_data.append(
            go.Scatter(
                x=brasil_argentina.index[:i+1],
                y=brasil_argentina[col].iloc[:i+1],
                mode='lines',
                name=col
            )
        )
    frames.append(go.Frame(data=frame_data))

fig.frames = frames

# Exibindo a figura
fig.show()
```
